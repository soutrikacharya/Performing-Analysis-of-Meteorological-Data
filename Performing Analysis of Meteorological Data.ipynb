```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('weatherHistory.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.fillna("NULL")
```

```python
df.dtypes
```

```python
df.describe()
```

```python
#Changing the datatype of Formatted Date column
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df['Formatted Date']
```

```python
df=df.set_index('Formatted Date')
df.head()
```

```python
df_monthly_mean = df[['Apparent Temperature (C)', 'Humidity']].resample('M').mean()
df_monthly_mean.head()
```

```python
#Plotting the variation in Apparent Temperature and Humidity with time

plt.figure(figsize=(15,5))
plt.title("Plot of Apparent Temperature and Humidity with time(in year)")
plt.plot(df_monthly_mean['Humidity'],label='Humidity',color='blue')
plt.plot(df_monthly_mean['Apparent Temperature (C)'],label='Apparent Temperature (C)',color='red')
plt.xlabel("Year",fontsize="16")
plt.legend(loc=2,fontsize=10)
```

```python
#Bar Plot

x=df_monthly_mean['Apparent Temperature (C)']
y=df_monthly_mean['Humidity']
fig= plt.figure(figsize=(10,5))
plt.bar(x,y, color ='blue', width = 0.4) 
  
plt.xlabel("Apparent Temperature") 
plt.ylabel("Humidity") 

plt.show()
```
